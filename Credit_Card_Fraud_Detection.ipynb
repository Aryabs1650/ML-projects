Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows of the dataset

credit_card_data.head()
```

```python
#last five rows of the dataset
credit_card_data.tail()
```

```python
#dataset informations
credit_card_data.info()
```

```python
#checking the number of missing values in each columns
credit_card_data.isnull().sum()
```

```python
#dealing with missing values

credit_card_data = credit_card_data.dropna(how='any')
```

```python
credit_card_data.isnull().sum()
```

```python
#distribution  of legit transaction and fradulent transactions
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0--> Normal Transaction

1--> Fraudulent Transaction

```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class ==0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both yransactions
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal and fraudulent transactions



No: of Fraudulent Transaction --> 419



```python
legit_sample=legit.sample(n=419)
```

Concatenating two Dataframe

```python
new_data=pd.concat([legit_sample,fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

Splitting the data

```python
X= new_data.drop(columns='Class',axis=1)
Y= new_data['Class']
```

To split the data into training data and testing data

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y , test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Model Training

Logistics Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

model evaluation

Accuracy Score

```python
X_train_pred= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_pred,Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_pred= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_pred,Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
